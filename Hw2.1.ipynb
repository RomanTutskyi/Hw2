```python
import pandas as pd
import numpy as np
tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Регіон", thousands=".", decimal=",")
table = tables[0]

table.replace("—", np.nan, inplace=True)


numeric_columns = table.columns[1:]

for column in numeric_columns:
    table[column] = pd.to_numeric(table[column], errors='coerce')


column_means = table.drop("Регіон", axis=1).mean()


df_filled = table.fillna(column_means)


average_birth_rate_2019 = df_filled["2019"].mean()

regions_above_average = df_filled[df_filled["2019"] > average_birth_rate_2019]["Регіон"]

print(regions_above_average)
```

```python
import pandas as pd
import numpy as np
tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Регіон", thousands=".", decimal=",")
table = tables[0]

table.replace("—", np.nan, inplace=True)

numeric_columns = table.columns[1:]

for column in numeric_columns:
    table[column] = pd.to_numeric(table[column], errors='coerce')

column_means = table.drop("Регіон", axis=1).mean()

df_filled = table.fillna(column_means)

region_highest = df_filled[df_filled["2014"] == df_filled["2014"].max()]["Регіон"].values[0]

print(region_highest)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

tables = pd.read_html("https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8", match="Регіон", thousands=".", decimal=",")
table = tables[0]

table.replace("—", np.nan, inplace=True)


numeric_columns = table.columns[1:]

for column in numeric_columns:
    table[column] = pd.to_numeric(table[column], errors='coerce')


column_means = table.drop("Регіон", axis=1).mean()


df_filled = table.fillna(column_means)


plt.figure(figsize=(10, 6))  


regions = df_filled["Регіон"]
birth_rates_2019 = df_filled["2019"]

# Побудова діаграми
plt.bar(regions, birth_rates_2019)
plt.xlabel("Регіони")
plt.ylabel("Народжуваність у 2019 році")
plt.title("Народжуваність по регіонах у 2019 році")
plt.xticks(rotation=90) 


plt.tight_layout()
plt.show()





```
