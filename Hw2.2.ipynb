```python
import pandas as pd

grouped = stats.groupby('Специализация')


aggregated_data = grouped['Зарплата.в.месяц'].agg(min_salary='min', max_salary='max')
aggregated_df = pd.DataFrame(aggregated_data)


def fill_avg_salary(row):
    return row.mean()

aggregated_df['avg'] = aggregated_df.apply(fill_avg_salary, axis=1) 


aggregated_df

```

```python
description = aggregated_df.describe()
print(description)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data=stats, x='Зарплата.в.месяц', bins=20, kde=True)
plt.xlabel('Заробітна плата')
plt.ylabel('Частота')
plt.title('Розподіл заробітних плат програмістів')
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(data=stats, x='Общий.опыт.работы', y='Зарплата.в.месяц')
plt.xlabel('Загальний досвід роботи (роки)')
plt.ylabel('Середня заробітна плата')
plt.title('Зв\'язок між досвідом роботи та заробітною платою')
plt.show()


plt.figure(figsize=(10, 6))
sns.countplot(data=stats, y='Специализация', order=stats['Специализация'].value_counts().index)
plt.xlabel('Кількість програмістів')
plt.ylabel('Спеціалізація')
plt.title('Кількість програмістів за спеціалізацією')
plt.show()
```
